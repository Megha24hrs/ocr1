```python
import cv2
import numpy as np

def normalize_image():
    # Read the image
    img = cv2.imread("/content/t10k-images.idx3-ubyte",cv2.IMREAD_GRAYSCALE)

    # Convert the image to floating-point representation
    img_float = np.float32(img)


    img_normalized = cv2.normalize(img_float, 0, 1, cv2.NORM_MINMAX)

    # Display the original and normalized images
    cv2.imshow('Original Image', img)
    cv2.imshow('Normalized Image', img_normalized)
    cv2.waitKey(0)


normalize_image()
```
